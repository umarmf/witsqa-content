## Imports

```python

import os
import sys
from datetime import datetime as dt
from datetime import date
from datetime import timedelta
import glob
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import calendar
```

```python
def remove_nonnumeric(string):
    if isinstance(string, str):
        string = ''.join(filter(lambda x: x.isdigit(), string))
    return string
```

```python
def combine_tiket_scrapes(depart_station, arrival_station):
    df = pd.DataFrame()
    files = glob.glob('..\scrape-csv\*.csv')
    route = ['\\' + depart_station + '-','-' + arrival_station + '-']

    for file in files:
        scrapedate = file[-12:-4]
        if all(x in file for x in route):
            past_df = pd.read_csv(file)
            past_df['scrape_date'] = pd.to_datetime(scrapedate)
            past_df['depart_station'] = depart_station
            past_df['arrival_station'] = arrival_station
            if df.shape == (0,0):
                df = past_df
            else:
                df = df.append(past_df)
    
    return df


# print('Earliest scrape: ' + str(min(df['scrape_datetime'])))
# print('Latest scrape: ' + str(max(df['scrape_datetime'])))
```

# Analysis assumptions, questions, & hypotheses
Let's get back to the objective of this analysis: 
When is the latest I can purchase an acceptable train ticket on each of those routes?

Assume the ticket will be purchased in a non-peak period.

Two key words here: "latest" and "acceptable".
Acceptable means the ticket fits my schedule and budget. Aka, a constraint.
Latest means how many days beforehand should I purchase the ticket. Aka, the objective.

The issue here is that "acceptable" is not set in stone. I may prefer Friday evenings on a budget train, but if those aren't available I may want to pay more or leave in a different time slot. Thus, I will take a "outside-in" approach: assume all tickets are "acceptable", then filter down to more preferable tickets. This outside-in approach allows me to get an overview of all the data before deciding to filter down.

Main questions to answer are:
1. When (in days) do tickets start becoming unavailable - main question
2. which times, days, classes are the least available - to guide filtering

Structure of analysis:
Analyze GMR-BDG, then PSE-KT tickets in each section

1) Data from one day of scraping:
a) Schedule
b) Availability

2) Data from 3+ days of scraping

3) Data from a month of scraping

Target variable: % unavailable tickets on a day/time/class 
= (Seats == 0) / (# seats in group)

##### Day x hour train routes
Are trains concentrated in certain time slots?

Calculate # tickets/day-hour = Count tickets in a day-hour/Occurences of each day-hour in the next 30 days

It seems that tickets can be divided into 5 groups:
Early morning (4-8), late morning (10), early afternoon (13-15), late afternoon (17-18), night (22-23)

```python
def dayhour_from_datetime(datetime_series):
    day_hour_series = datetime_series.dt.day_name().str.slice(0,3,1) + datetime_series.dt.strftime('%H')
    return day_hour_series
```

```python
# generate list of 30 days from 24jan to 24feb
check_date_plus30 = (dt.strptime(check_date, "%Y-%m-%d")+ timedelta(days=30)).strftime("%Y-%m-%d")
date_range_30days = pd.date_range(check_date, check_date_plus30, periods=30*24+1).to_series() #.reset_index()
date_range_30days_df = pd.DataFrame({'date': date_range_30days, 'day_name': date_range_30days.dt.day_name(), 'day': date_range_30days.dt.dayofweek, 'hour': date_range_30days.dt.hour })
dayhour_30days = date_range_30days_df.groupby(['day_name', 'day','hour']).size().reset_index().rename(columns={0: 'count_dayhour'})
```

```python
df_one_bd_dayhour = pd.DataFrame({'day': df_one_bd.depart_datetime.dt.dayofweek, 'hour': df_one_bd.depart_datetime.dt.hour })
df_one_bd_dayhour = df_one_bd_dayhour.groupby(['day','hour']).size().reset_index().rename(columns={0: 'count_routes'})
df_one_bd_dayhour = dayhour_30days.merge(df_one_bd_dayhour, how='left', on =['day','hour'])
df_one_bd_dayhour['routes_per_dayhour'] = df_one_bd_dayhour['count_routes']/df_one_bd_dayhour['count_dayhour']
df_one_bd_dayhour.fillna(0, inplace=True)

# add dayhour column and order by day, hour

df_one_bd_dayhour['dayhour'] = df_one_bd_dayhour['day_name'].str.slice(0,3,1) + '-' + df_one_bd_dayhour['hour'].astype('str')
df_one_bd_dayhour['x_axis'] = df_one_bd_dayhour.apply(lambda x : x['dayhour'] if str(x['hour']) in ['0','8','16'] else '', axis=1)
df_one_bd_dayhour.sort_values(['day','hour'], inplace=True)

plt.figure(figsize=(22,4))
plt.plot('dayhour', 'routes_per_dayhour', data = df_one_bd_dayhour)
plt.xticks(df_one_bd_dayhour['x_axis'])
plt.grid(b=True, which='both', color='0.65', linestyle='-')
plt.show()
```
